```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
```

```python
def getAllDoxyDonkeyPosts(url, links):
    response = urllib.request.urlopen(url)
    soup = BeautifulSoup(response, "lxml")
    for a in soup.findAll('a'):
        try:
            url = a['href']
            title = a['title']
            if title == "Older Posts":
                print(title, url)
                links.append(url)
                getAllDoxyDonkeyPosts(url, links)
        except:
            title = ""
    return
```

```python
links = []
url = "http://doxydonkey.blogspot.in"

getAllDoxyDonkeyPosts(url, links)
```

```python
def getDoxyDonkeyText(url):
    response = urllib.request.urlopen(url)
    soup = BeautifulSoup(response, "lxml")
    posts = []
    
    for div in soup.findAll("div", {"class": "post-body"}):
        posts += map(lambda p: p.text.encode('utf-8'), div.findAll("li"))
    return posts
```

```python
posts  = []

for link in links:
    posts += getDoxyDonkeyText(link)
len(posts)
```
